# Charts

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
house_prices = pd.read_csv("data/house-prices.csv")
````

## Histogram

### Plotting the series

````python
sns.displot(house_prices["SalePrice"])
plt.title("Sale Price Distribution")
plt.xlabel("Sale Price")
plt.ylabel("Houses Sold")
plt.show()
````

### Styled

````python
sns.set_style("white")
sns.set_color_codes(palette='deep')
f, ax = plt.subplots(figsize=(8, 7))
sns.histplot(house_prices['SalePrice'], color="b", kde=True)
ax.xaxis.grid(False)
ax.set(ylabel="Houses Sold")
ax.set(xlabel="Sale Price")
ax.set(title="Sale Price Distribution")
sns.despine(trim=True, left=True)
plt.show()
````

### Plotting using the data frame

````python
sns.displot(house_prices, x="SalePrice")
plt.title("Sale Price Distribution")
plt.xlabel("Sale Price")
plt.ylabel("Houses Sold")
plt.show()
````

### Adjusting the figure size

````python
fig, ax = plt.subplots(figsize=(3, 2))
# Note that displot doesn't accept an `ax` parameter 
# which is why we're using histplot here
sns.histplot(house_prices, x="SalePrice", ax=ax)
plt.show()
````

## Scatter Plot

### Seaborn

````python
sns.scatterplot(data=house_prices, x="TotalBsmtSF", y="SalePrice")
plt.show()
````

## Pyplot

````python
house_prices.plot.scatter(x="TotalBsmtSF", y="SalePrice", c="DarkBlue")
plt.show()
````

## Box Plot

````python
sns.boxplot(data=house_prices, x="OverallQual", y="SalePrice")
plt.show()
````

## Bar plot

"Show point estimates and confidence intervals as rectangular bars."

````python
sns.barplot(data=house_prices, x="MoSold", y="SalePrice")
plt.title("Average Sale Price by Month")
plt.ylabel("Average Sale Price")
plt.xlabel("Month Sold")
plt.show()
````

````python
percent_missing = house_prices.isnull().sum() / house_prices.shape[0]
features_with_missing_values = percent_missing[percent_missing.gt(0)].mul(100).sort_values(ascending=False)
sns.barplot(x=features_with_missing_values.index, y=features_with_missing_values)
plt.xticks(rotation=90)
plt.ylabel("Percent of Values Missing")
plt.show()
````

## Correlation Heatmap

See Correlation notebook for more detailed examples.

````python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(house_prices.corr());
plt.show()
````
