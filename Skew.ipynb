# Skew

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

## Plotting the skewed data

```python
df = pd.read_csv("data/house-prices.csv")
sale_price = df["SalePrice"]

sns.displot(sale_price)
plt.show()
```

## Applying transformations to remove the skew

### Log transformation

```python
sns.displot(np.log(sale_price))
plt.show()
```

### Square root

```python
sns.displot(np.sqrt(sale_price))
plt.show()
```

### Reciprocal

```python
sns.displot(1 / sale_price)
plt.show()
```

## Measuring skewness

```python
df.skew().sort_values(ascending=False)
```

```python
feature_skew = df.select_dtypes(include=np.number).skew()
log_features = feature_skew[abs(feature_skew) > 0.9].index
log_features
```

## Kurtosis

```python
df.kurt().sort_values(ascending=False)
```

## Probability plots

For skewed distributions, this won't be a straight line:

```python
res = stats.probplot(sale_price, plot=plt)
plt.show()
```

Normal distributions will be closer to a straight line:

```python
res = stats.probplot(np.log(sale_price), plot=plt)
plt.show()
```
