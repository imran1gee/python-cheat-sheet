# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("data/titantic-train.csv")

display(df.head())
```

## Simple Pair Plot

```python
subset = df[["Age", "Fare"]]
sns.pairplot(subset)
plt.show()
```

## Segmented Pair Plot

```python
subset = df[["Age", "Fare", "Sex"]]
sns.pairplot(subset, hue="Sex")
plt.show()
```

## Segmented Pair Plot with Histogram along Diagnol

```python
subset = df[["Age", "Fare", "Sex"]]
sns.pairplot(subset, hue="Sex", diag_kind="hist")
plt.show()
```
